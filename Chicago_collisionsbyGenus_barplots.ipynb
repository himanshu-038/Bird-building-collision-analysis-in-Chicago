```python
#importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing Bird collision data from Chicago
Chicago_collisions = pd.read_csv("Bird_building_collisions\Chicago_collision_data.csv")
```

```python
#Exploring the data
Chicago_collisions.info()
Chicago_collisions.shape
Chicago_collisions.head()
```

```python
#Checking for NaNs in the dataframe
Chicago_collisions.isnull().values.any()
Chicago_collisions.isnull().sum()
```

```python
#Checking the range of data in Date column
print(Chicago_collisions["Date"].min())
print(Chicago_collisions["Date"].max())
```

```python
#Making dataframe of the number of collisions recorded for each genus
collisions_by_genus = Chicago_collisions["Genus"].value_counts()
Ch_collisions_by_genus = pd.DataFrame({"genus":collisions_by_genus.index, "collisions":collisions_by_genus.values})
```

```python
#using seaborn to plot the count of collisions recorded for the categorical variable genus
plt.figure(figsize = (10,20))
sns.catplot(data = Chicago_collisions, y = "Genus", kind = "count")
plt.show()
```

```python
# Making bar chart of the collisions by genus using matplotlib
plt.figure(figsize=(8,10))
plt.ylabel("Count")
plt.barh(Ch_collisions_by_genus["genus"],Ch_collisions_by_genus["collisions"])
plt.show()
```
